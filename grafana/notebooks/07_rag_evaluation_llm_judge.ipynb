```python
from pathlib import Path
import matplotlib.pyplot as plt
from evaluation.llm_judge import llm_judge

result = llm_judge("GDPR?", "EU summary", context="stub")
print(result)

names = [c.name for c in result.criteria]
scores = [c.score for c in result.criteria]

out_dir = PROJECT_ROOT / "images"
out_dir.mkdir(parents=True, exist_ok=True)
out = out_dir / "rag_judge_scores.png"

if names:
    fig, ax = plt.subplots(figsize=(12, 6))
    ax.bar(names, scores)
    ax.set_title("RAG Judge Criteria Scores")
    ax.set_ylabel("Score")
    ax.set_ylim(0, 5)
    ax.tick_params(axis="x", rotation=45)
    fig.tight_layout()
    fig.savefig(out, dpi=150)
    plt.close(fig)
    print(f"Saved: {out}")
else:
    print("No criteria returned by judge; PNG not created.")
```
