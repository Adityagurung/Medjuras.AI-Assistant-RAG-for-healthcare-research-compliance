# Notebook 5: Hybrid RRF evaluation

Fuses Qdrant dense (2.0) and Elasticsearch BM25 (1.0) with RRF k=60.


```python
import _bootstrap  # noqa: F401
import json
from evaluation.config_paths import load_ground_truth
from evaluation.eval_utils import evaluate
from evaluation.results_utils import results_path, results_images_dir
from evaluation.retrieval_results_charts import save_single_method_plot, save_comparison_from_result_files
from ingestion.paths import RESULTS_DIR
from search.hybrid_search import hybrid_search

TOP_K = 10
gt = load_ground_truth()
gt_rows = [{'query': r['question'], 'doc_id': r['doc_id']} for r in gt]
metrics = evaluate(gt_rows, lambda q, **kw: hybrid_search(q, top_k=TOP_K, local=True), top_k=TOP_K)
out = results_path('hybrid_rrf_metrics.json')
out.write_text(json.dumps(metrics, indent=2), encoding='utf-8')
png = save_single_method_plot('Hybrid', metrics, TOP_K, results_images_dir() / 'hybrid_rrf_metrics.png')
comparison_png = save_comparison_from_result_files(RESULTS_DIR, TOP_K)
print(metrics)
print('Metrics JSON:', out.resolve())
print('Chart PNG:', png.resolve())
print('Comparison Chart PNG:', comparison_png.resolve())
```
