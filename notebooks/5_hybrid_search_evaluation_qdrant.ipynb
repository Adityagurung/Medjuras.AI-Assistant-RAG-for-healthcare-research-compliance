# Notebook 5: Hybrid RRF evaluation

Fuses Qdrant dense (2.0) and Elasticsearch BM25 (1.0) with RRF k=60.


```python
import _bootstrap  # noqa: F401
import json
from evaluation.config_paths import ground_truth_path
from evaluation.eval_utils import evaluate
from evaluation.results_utils import results_path
from search.hybrid_search import hybrid_search

gt = json.loads(ground_truth_path().read_text(encoding='utf-8'))
gt_eval = [{'query': r['question'], 'doc_id': r['doc_id']} for r in gt]
metrics = evaluate(gt_eval, lambda q, **kw: hybrid_search(q, top_k=10, local=True), top_k=10)
out = results_path('hybrid_rrf_metrics.json')
out.write_text(json.dumps(metrics, indent=2), encoding='utf-8')
print(metrics)
```
