# Notebook 3: Keyword / BM25 evaluation (Elasticsearch)

Measures Hit@K, MRR, and nDCG using **Elasticsearch BM25**.


```python
import _bootstrap  # noqa: F401
import json
from evaluation.config_paths import load_ground_truth
from evaluation.eval_utils import evaluate
from evaluation.results_utils import results_path, results_images_dir
from evaluation.retrieval_results_charts import save_single_method_plot
from search.es_search import search_elasticsearch

TOP_K = 10
gt = load_ground_truth()
gt_rows = [{'query': r['question'], 'doc_id': r['doc_id']} for r in gt]
metrics = evaluate(gt_rows, lambda q, **kw: search_elasticsearch(q, top_k=TOP_K, local=True), top_k=TOP_K)
out = results_path('es_bm25_metrics.json')
out.write_text(json.dumps(metrics, indent=2), encoding='utf-8')
png = save_single_method_plot('Elasticsearch', metrics, TOP_K, results_images_dir() / 'es_bm25_metrics.png')
print(metrics)
print('Saved chart:', png)
```
