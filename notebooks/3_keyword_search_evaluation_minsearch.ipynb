# Notebook 3: Keyword / BM25 evaluation (Elasticsearch)

Measures Hit@K, MRR, MAP, nDCG using **Elasticsearch BM25**.


```python
import _bootstrap  # noqa: F401
import json
from evaluation.config_paths import load_ground_truth
from evaluation.eval_utils import evaluate
from evaluation.results_utils import results_path
from search.es_search import search_elasticsearch

gt = load_ground_truth()
gt_eval = [{'query': r['question'], 'doc_id': r['doc_id']} for r in gt]
metrics = evaluate(gt_eval, lambda q, **kw: search_elasticsearch(q, top_k=10, local=True), top_k=10)
out = results_path('es_bm25_metrics.json')
out.write_text(json.dumps(metrics, indent=2), encoding='utf-8')
print(metrics)
```
